```python
# create a flag column using an aggregated criteria through transform()
# future warning: transform(pd.max) will turn into transform('max') in the future. Recommended to use transform('max') instead.

df_orders_products_merge['max_order'] = df_orders_products_merge.groupby(['user_id'])['customer_order_count'].transform('max')
```

```python
# check functionality
df_orders_products_merge.head()
```

```python
# declares empty list that will fill the 'loyalty_flag' column
flag_values_list = []

# for loop that fills the empty list 
for value in df_orders_products_merge['max_order']:  
    if value > 40:
        flag_values_list.append('Loyal customer')
    elif value <= 40 and value > 10:
        flag_values_list.append('Regular customer')
    else:
        flag_values_list.append('New customer')
```

```python
# create a column in the dataframe. Values come from "flag_values_list"
df_orders_products_merge['loyalty_flag'] = flag_values_list
```

### 6. Creating a spending flag for customers

```python
# create a column using aggregation through transform(). Mean aggregation of 'prices'. Groupby 'user_id'.

df_orders_products_merge['mean_price'] = df_orders_products_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
# check functionality
df_orders_products_merge.head(40)
```

```python
# create a flag column depeding on the mean price. use .loc 
# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a Low spender.
# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a High spender.

df_orders_products_merge.loc[df_orders_products_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
df_orders_products_merge.loc[df_orders_products_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# check functionality through value_counts()
df_orders_products_merge['spending_flag'].value_counts()
```

### 7. Creating a frequency flag column for customers

```python
# create an aggregated column using tranform('median'). groupby()
df_orders_products_merge['median_prior_days'] = df_orders_products_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
# check functionality 
df_orders_products_merge.head(1)
```

```python
# create a flag column with the following conditions 
# If the median of days_since_prior_order is higher than 20, then the customer should be labeled a Non-frequent customer
# If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a Regular customer
# If the median is lower than or equal to 10, then the customer should be labeled a Frequent customer

df_orders_products_merge.loc[df_orders_products_merge['median_prior_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
df_orders_products_merge.loc[(df_orders_products_merge['median_prior_days'] <= 20) & (df_orders_products_merge['median_prior_days'] > 10), 'frequency_flag'] = 'Regular customer'
df_orders_products_merge.loc[df_orders_products_merge['median_prior_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
