### Importing Libraries

```python
import pandas as pd
import numpy as np
import os
```

### Importing Data

```python
# establishes data path
data_path = r'C:\Users\gerar\CareerFoundry Python\Instacart Basket Analysis\02 Data'
```

```python
# imports pickle file into a panda dataframe 
df_orders_products_merge = pd.read_pickle(os.path.join(data_path, 'Prepared Data', 'orders_products_merge_busy.pkl'))
```

```python
# creates test dataframe
df_test = df_orders_products_merge[:1000000]
```

```python
df_test.head(1)
```

```python
# uses grouby function. This only groups data according to a specific column
df_test.groupby('department_id')
```

```python
# use  groupby().agg() for a single aggregation (BEST PRACTICE)
df_test.groupby('department_id').agg({'customer_order_count': ['mean']})
```

```python
# alternative way to calculate mean (not best practice)
df_test.groupby('department_id')['customer_order_count'].mean()
```

```python
# calculating multiple aggregations
df_test.groupby('department_id').agg({'customer_order_count' : ['mean', 'min', 'max']})
```

```python
df_test.head(100)
```

```python
# create a flag column using an aggregated criteria through transform()
df_test['max_order'] = df_test.groupby(['user_id'])['customer_order_count'].transform(np.max)
```

```python
df_test.head(100)
```

```python
pd.options.display.max_rows = None
```

```python
df_test.head(100)
```

```python
# declares empty list that will fill the 'busiest_period_of_day' column
list_flag_values = []

# for loop that fills the empty list 
for value in df_test['max_order']:  
    if value > 40:
        list_flag_values.append('Loyal customer')
    elif value <= 40 and value > 10:
        list_flag_values.append('Regular customer')
    else:
        list_flag_values.append('New customer')
```

```python
df_test['loyalty_flag'] = list_flag_values 
```
