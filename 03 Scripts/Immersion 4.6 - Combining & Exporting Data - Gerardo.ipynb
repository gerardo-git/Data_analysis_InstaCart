### 5. Combining data frames 

```python
# Checking keys of both dataframes
df_products_consistent.head(1)
```

```python
# the last column "_merge" is causing an error when merging, it needs to be deleted
df_orders_products_combined
```

```python
# getting rid of '_merge' column to avoid ValueError. 
df_orders_products_combined = df_orders_products_combined.drop('_merge', axis=1)
```

```python
# merging df_products_consistent and df_orders_products_combined
df_orders_products_merge = df_products_consistent.merge(df_orders_products_combined, on = ['product_id'], indicator = True ) 
```

```python
df_orders_products_merge
```

### 6. Confirming merge results

```python
# value counts of merge flag
df_orders_products_merge['_merge'].value_counts()
```

### 7. Exporting merged dataframe
